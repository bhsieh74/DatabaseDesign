I renamed the columns, removing the spaces and parentheses so that performing analysis and visualization in the latter part of this assignment would be less messy and convenient. Then, I made all column names lowercase to keep consistent.

This visualization builds upon one of the analyses that I conducted in the earlier section. In this visualization, I group the dataframe by beverage_category and use the aggregate mean function. Then, I single out the calories variable to use in my bar chart. Then, I find the unique beverage category names and sort them alphabetically and then convert them to a Pandas Series to use as my labels in my bar chart. I have to sort it alphabetically otherwise the bar chart will not have the correct average calories. 

```python
mean = df.groupby(['beverage_category']).mean()
mean = mean['calories']
labels = pd.Series(df.beverage_category.unique()).sort_values()
plt.barh(labels, mean)
plt.xlabel("Calories")
plt.ylabel("Beverage Category")
plt.title("Average Calories per Beverage Category")
```

In this visualization, I use a scatter plot. First, I group the dataframe by calories and use the mean aggregate function upon it. Then, I create a new Pandas Series named averageprotein and assign it to the protein_g column of the new dataframe. Next, I created a new variable named indexes that contains the index of the new dataframe, which consists of all the different calorie values. Finally, I can plug these two values as the parameter for my scatter plot. 

```python
calories = df.groupby(['calories']).mean()
averageprotein = calories['protein_g']
indexes = calories.index
plt.scatter(indexes, averageprotein)
plt.xlabel("Calories")
plt.ylabel("Average Grams of Protein")
plt.title("Average Grams of Protein based on Calories in a Starbucks Drink")
```

This visualization shows the proportion of beverage categories out of all the Starbucks beverages. I group the dataframe by beverage category and use the count aggregate function upon it. Then, I assign a variable named labels to the index of this new dataframe to get all of the different beverage categories. Then, I assign a new variable count to get the count of each beverage category and input them as the parameters of my pie chart. As we can see, plain coffee is the least frequent Starbucks drink. 

```python
beveragecategories = df.groupby(['beverage_category']).count()
labels = beveragecategories.index
count = beveragecategories['beverage']
plt.pie(count, labels=labels, autopct='%1.1f%%')
plt.show()
```
